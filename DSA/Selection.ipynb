```python
import random

def bubble_sort(A):
    for i in range(len(A)-1):
        for j in range(len(A)-1):
            if A[j]>A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
            
    return A



def quick_select(A, low, hi, k):
    if low == hi:
        return A[low]
    
    split = partition_1(A, low, hi)
    length = split - low + 1
    
    if length == k:
        return A[split]
    elif k < length:
        return quick_select(A, low, split-1, k)
    else:
        return quick_select(A, split+1, hi, k-length)
    
def get_pivot(A, low, hi): # choose median value of low mid hi values as pivot
    mid = (hi + low) // 2
    pivot = hi
    if A[low] < A[mid]:
        if A[mid] < A[hi]:
            pivot = mid
    elif A[low] < A [hi]:
        pivot = low
    
    return pivot

    
def partition_1(A, low, hi): # partition_1 is better than partition_2, because it choose median value as pivot.
    pivotIndex = get_pivot(A, low, hi) # make sure to get time O(nlogn)
    pivotValue = A[pivotIndex]
    A[pivotIndex], A[low] = A[low], A[pivotIndex]
    border = low
    
    for i in range(low, hi+1): # use one pointer: border, and a for loop to traverse whole list
        if A[i] < pivotValue:
            border += 1
            A[i], A[border] = A[border], A[i]
    A[low], A[border] = A[border], A[low]
    
    return border
        
def partition_2(A, low, hi):# choose first element as pivot gives worse case time O(n2) if input array is sorted
    pivotvalue = A[low]
    LM = low + 1
    RM = hi
    
    while True:
        while LM < hi and A[LM] < pivotvalue:
            LM += 1
        while RM > low and A[RM] > pivotvalue:
            RM -= 1
        if LM >= RM:
            break
        else:
            A[LM], A[RM] = A[RM], A[LM]
    
    A[low], A[RM] = A[RM], A[low]
    
    return RM

def main():
    list = []
    for x in range(10):
        list.append(random.randint(1,100))
    
    print(list)
    print(bubble_sort(list))
    print(quick_select(list, 0, len(list)-1, 4))
    
if __name__ == "__main__":
    main()


        
```

```python
import random

def bubble_sort(A):
    for i in range(len(A)-1):
        for j in range(len(A)-1):
            if A[j]>A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
            
    return A

def quick_select(A, low, hi, k):
    if low == hi:
        return A[low]
    
    split = partition_3(A, low, hi)
    length = split - low + 1
    
    if length == k:
        return A[split]
    elif k < length:
        return quick_select(A, low, split-1, k)
    else:
        return quick_select(A, split+1, hi, k-length)
    
def median_of_medians(A): #get pivot value, approximate median value of the list
    sub_A = [[A[j:j+5]] for j in range(0, len(A), 5)]
  
    medians = []
    for item in sub_A:
        medians.append(sorted(item[0])[len(item)//2])
        
    if len(medians) <= 5:
        return sorted(medians)[len(medians)//2]
    else:
        return median_of_medians(medians)
    
def partition_3(A, low, hi): #use approximate median as pivot to get time O(n)
    pivotValue = median_of_medians(A)
    pivotIndex = A.index(pivotValue)
    A[pivotIndex], A[low] = A[low], A[pivotIndex]
    border = low
    
    for i in range(low, hi+1): # use one pointer: border, and a for loop to traverse whole list
        if A[i] < pivotValue:
            border += 1
            A[i], A[border] = A[border], A[i]
    A[low], A[border] = A[border], A[low]
    
    return border


def main():
    list = []
    for x in range(10):
        list.append(random.randint(1,100))
    
    print(list)
    print(bubble_sort(list))
    print(quick_select(list, 0, len(list)-1, 4))
    
if __name__ == "__main__":
    main()
```
